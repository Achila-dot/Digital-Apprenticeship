```python
List_dict=[]
for item in set(relevant_data['communities_villages'].values):
    It_em=relevant_data.loc[df['communities_villages']==item]
    It_em_count=It_em.groupby(['water_functioning']).count()
    Dict=It_em_count.to_dict()
    Dict[item] = Dict.pop('communities_villages')
    data_c=pd.DataFrame.from_dict(Dict,orient='index')
    List_dict.append(data_c)
Relevant_Frame=pd.concat(List_dict)
```

```python
Relevant_Frame=Relevant_Frame.fillna(0)
Relevant_Frame.rename(columns={'yes':'Functional','no':'Broken'},inplace=True)
Relevant_Frame['water_points']=Relevant_Frame.sum(axis=1)
Relevant_Frame.drop('na_dn',axis='columns',inplace=True)
Relevant_Frame['community_ranking']=(Relevant_Frame['Functional']/Relevant_Frame['water_points'])*100
Relevant_Frame.sort_values(by='community_ranking',ascending=False,inplace=True)
Relevant_Frame.head()
```
